```python
import cv2
import numpy as np
from urllib.request import urlopen
from matplotlib import pyplot as plt

def read_image_from_url(url):
    """Reads an image from a URL and returns it as a BGR numpy array."""
    response = urlopen(url)
    image = np.asarray(bytearray(response.read()), dtype="uint8")
    image = cv2.imdecode(image, cv2.IMREAD_COLOR)
    return image

# Example image URL
url = 'https://hips.hearstapps.com/hmg-prod/images/403907084-684760070295678-4015199202133812968-n-655c4ce2b26a8.jpg?crop=1xw:1xh;center,top&resize=980:*'
# Read the image from the URL
img_Winter_bgr = read_image_from_url(url)

# Height, Width, and Three Channels
print('Image Size',img_Winter_bgr.shape)

# Splitting the BGR image into its components
b, g, r = cv2.split(img_Winter_bgr)

# Merging the components into an RGB image
img_Winter_rgb = cv2.merge([r, g, b])

# Setting up the matplotlib plot
fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(12, 4))

def f_imshow(ax, mat, title):
    ax.imshow(mat, cmap='PuRd_r')
    ax.set_title(title)
    ax.axis('off')

# Display each channel and the combined image
f_imshow(axs[0], img_Winter_rgb, 'All Channels(RGB)')
f_imshow(axs[4], img_Winter_bgr, 'Original Channels(BGR)' )
f_imshow(axs[1], r, 'Red Channel')
f_imshow(axs[2], g, 'Green Channel')
f_imshow(axs[3], b, 'Blue Channel')

# Adjust the layout of the subplots
plt.tight_layout(w_pad=0.5)
plt.show()
```
